```python
import pandas as pd
from sklearn.mixture import GaussianMixture
from sklearn import svm
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.metrics import roc_auc_score

df = pd.read_csv('SRP158491_converted.tsv', sep='\t')

X = df.drop(columns=['Gene']).T

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

pca = PCA(n_components=50, random_state=42)
X_reduced = pca.fit_transform(X_scaled)

n_clusters = 3
gmm = GaussianMixture(n_components=n_clusters, reg_covar=1e-3, random_state=42)
clusters = gmm.fit_predict(X_reduced)  # Assign cluster labels

y = clusters  # Cluster labels

X_train, X_test, y_train, y_test = train_test_split(X_reduced, y, test_size=0.25, random_state=42)

kernels = ['linear', 'poly', 'rbf', 'sigmoid']
for kernel in kernels:
    print(f"Kernel: {kernel}")
    
    model = svm.SVC(kernel=kernel,  probability=True, random_state=42)
    model.fit(X_train, y_train)
    predicted = model.predict(X_test)
    
    score = accuracy_score(y_test, predicted)
    predicted_probs = model.predict_proba(X_test)
    
    roc_auc = roc_auc_score(y_test, predicted_probs, multi_class='ovr', average='macro')
    
    print(f"Accuracy Score with {kernel} kernel: {score}\n")
    print(f"ROC Accuracy Score with {kernel} kernel: {roc_auc}\n")
```

```python
import pandas as pd
from sklearn.mixture import GaussianMixture
from sklearn import svm
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

df = pd.read_csv('SRP158491_converted.tsv', sep='\t')

def get_top_genes_by_variance(df, num_genes):
    df_var = df.set_index('Gene').var(axis=1)
    top_genes = df_var.nlargest(num_genes).index
    return df.set_index('Gene').loc[top_genes].T.dropna().astype('float32')

num_genes_list = [10, 100, 1000, 10000]
results = {}

for num_genes in num_genes_list:
    print(f"\nTraining with top {num_genes} genes...")
    
    try:
        X = get_top_genes_by_variance(df, num_genes)
    except MemoryError:
        print(f"MemoryError: Unable to process {num_genes} genes.")
        continue
    
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)
    
    if num_genes > 100:
        pca = PCA(n_components=100, random_state=42)  # Adjust components as needed
        X_scaled = pca.fit_transform(X_scaled)
        print(f"  Reduced dimensionality with PCA to {X_scaled.shape[1]} components")
    
    n_clusters = 3
    gmm = GaussianMixture(
        n_components=n_clusters,
        covariance_type="diag",
        reg_covar=1e-2,
        random_state=42
    )
    try:
        clusters = gmm.fit_predict(X_scaled)
    except ValueError:
        print(f"Failed to fit GMM for {num_genes} genes with current configuration.")
        continue
    
    X_train, X_test, y_train, y_test = train_test_split(X_scaled, clusters, test_size=0.25, random_state=42)
    
    kernels = ['linear', 'rbf']
    for kernel in kernels:
        print(f"  Using kernel: {kernel}")
        
        model = svm.SVC(kernel=kernel, random_state=42)
        model.fit(X_train, y_train)
        predicted = model.predict(X_test)
        
        score = accuracy_score(y_test, predicted)
        print(f"    Accuracy Score with {kernel} kernel: {score}")
        
        results[f"{num_genes}_genes_{kernel}"] = score
```
